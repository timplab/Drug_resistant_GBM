```python
import pandas as pd
from functools import reduce
import seaborn as sns
import matplotlib.pyplot as plt
```

```python
df1 = pd.read_csv("flair_glio_SE_events_42MGBAres_v_42MGBA_fishersP.tsv",delimiter="\t",usecols=["event","42MGBA_42MGBAres.adj.pval"])
df2 = pd.read_csv("flair_glio_SE_events_T98G_v_42MGBA_fishersP.tsv",delimiter="\t",usecols=["event","T98G_42MGBA.adj.pval"])
df3 = pd.read_csv("flair_glio_SE_events_T98G_v_42MGBAres_fishersP.tsv",delimiter="\t", usecols=["event","T98G_42MGBAres.adj.pval"])
data_frames = [df1, df2, df3]
df_merged = reduce(lambda  left,right: pd.merge(left,right,on=['event'],how='inner'), data_frames)
```

```python
df_merged.set_index("event",inplace=True)
df_merged.head()
```

```python
fig = plt.figure(figsize=(12,12))
r = sns.heatmap(df_merged, cmap='BuPu')
```
